# Задача 1

На вход задачи подаются $2n$ чисел. Постройте алгоритм, который разбивает их на пары так, что максимальная из сумм чисел в парах минимальна.

```python
#Пример входного массива. Должно работать на любом числовом массиве.
input_ = [1, 2, 7, 8, 9, 12, 56, 76, 89, 4, 5, 6, 8, 1] 

test_cases = [
    input_,
    [],
    [1, 1],
    [0, -1, 0, 0]
]
#Ваш код тут
def ft_pairs(seq: list) -> list:
    res = []
    x = len(seq) // 2
    sorted_seq = sorted(seq)
    
    for i in range(x):
        res.append((sorted_seq[i], sorted_seq[-i-1]))
        
    return res
    
#В качестве ответа выведите распечатку составленных пар чисел
for test in test_cases:
    print(ft_pairs(test))
```

# Задача 2

Пусть $X$ это набор из $N$ точек из $R^2$. Предполагаем, что все точки имеют различные координаты, т.е. для любой точки $(x,y)$ и $(x′,y′)$ из $X$ мы гарантируем, что $x\neq x'$ и $y\neq y′$. Назовем точку $(x,y)$ из $X$ Парето-оптимальной, если не существует другой точки $(x′,y′)$ из $X$ такой, что $x′<x$ и $y′<y$. Выпишите все Парето-оптимальные точки из $X$, используя асимптотически не более $O(NlogN)$ сравнений координат.

```python
#Пример входного массива. Должно работать на любом, удовлетворяющем условиям выше.
X = [(2, 2), (3, 3), (4, 1), (5, 6), (1, 4), (6, 5)] 

test_cases = [
    X,
    [(1, 1), (5, 7), (4, 0), (2, 8), (-5, 20)],
    [(0, 0)],
    [(0, 1), (1, 0)],
    [],
    [(1, 1), (1, 2)],
    [(10, -10), (0, 0), (10, 0), (0, 10)],
    [(10, -10), (0, 0), (10, 0), (0, -10)],
    [(1, 2), (2, 2), (2, 1), (1, 1)],
    [(100, 100), (2, 2), (50, 0), (1, 1)]
]
#Ваш код тут

# Простое решение, на основе которого будем валидировать более сложную реализацию
# O(n^2)
def find_points(seq: list) -> list:
    res = []
    k = 0
    
    for x in seq:       
        for x_ in seq:
            if x != x_:
                if x[0] < x_[0] or x[1] < x_[1]:
                    pass
                else:
                    k += 1
                    break
        
        if k == 0:
            res.append(x)
        
        k = 0
    
    return res

# Потенциально правильное решение
# O (n log n)
def find_points_corr(seq: list) -> list:
    # создаем массив точек и записываем в него "плохие" точки
    point_list = [0 for x in range(len(seq))]
    
    # результирующий массив
    res = []
    
    # к - костыль :) на самом деле это переменная, которая проверяет что точка подходит под условия
    k = 0
    
    # идём от первого элемента до предпоследнего
    for i, x in enumerate(seq[:-1]):
        
        # если мы забраковали данную точку то пропускаем её
        if point_list[i] == 0:
            
            # проверяем начиная со следующего элемента до последнего
            for j, x_ in enumerate(seq[i+1:]):
                
                # если точка удовлетворяет нашим условиям
                if x[0] < x_[0] or x[1] < x_[1]:
                    
                    # то проверяем является ли вторая точка 
                    # "плохой" - т.е. такой, что впоследствии её не нужно будет проверять
                    
                    ## тут ОБЯЗАТЕЛЬНО нестрогое неравенство иначе будет ошибка на 7-ом тесте
                    if x[0] <= x_[0] and x[1] <= x_[1]: 
                        
                        # если точка "плохая" - увеличиваем номер элемента в списке
                        point_list[j+i+1] += 1
                        
                # если точка не удовлетворяет условиям - инкрементируем костыль
                else:
                    k += 1

            if k == 0:
                res.append(x)
        
            k = 0

    # добавляем проверку пустого списка
    if len(seq) > 0:
        
        # если последняя точка не является плохой то добавляем её в результирующий список
        if point_list[-1] == 0:            
            res.append(seq[-1])
        
    return res


#В качестве ответа выведите координаты искомых точек
for test in test_cases:
    print(find_points(test))
    print(find_points(test) == find_points_corr(test))
    print()
    
```
